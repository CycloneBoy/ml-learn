```python
import pandas as pd
import xgboost as xgb
import numpy as np

from util.constant import CONCRETE_DATA_DIR

data = pd.read_csv("./Concrete_Data.csv",encoding='unicode_escape')
data.head(10)

```

```python

data.rename(columns={"Concrete compressive strength(MPa, megapascals) ":'label'},inplace=True)
data.head(10)



```

```python
mask = np.random.rand(len(data)) < 0.8
train = data[mask]
print(mask)
test = data[~mask]

xgb_train = xgb.DMatrix(train.iloc[:, :7], label=train.label)
xgb_test = xgb.DMatrix(test.iloc[:, :7], label=test.label)

```

```python

params = {
    "objective": "reg:linear",
    "booster": "gbtree",
    "eta": "0.1",
    "min_child_weight": 1,
    "max_depth": 5,
}


num_round = 50
watchlist = [(xgb_train, 'train'), (xgb_test, 'test')]

bst = xgb.train(params, xgb_train, num_round, watchlist)
bst.save_model('./concrete.model')

```

```python
bst = xgb.Booster()
bst.load_model('./concrete.model')
pred = bst.predict(xgb_test)
print(pred)
```
